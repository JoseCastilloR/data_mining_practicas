```python
# Imports

import pandas as pd
import io

from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['metacritic_games.csv']))
df.head()
```

```python
to_drop = ['number_players', 'positive_critics', 'neutral_critics', 'negative_critics', 'positive_users', 'neutral_users', 'negative_users']

df.drop(to_drop, inplace=True, axis=1)
df.head()
```

```python
df = df.drop(labels=range(len(df))[5000:], axis=0)
print("Number of rows present: ", len(df))
```

```python
df.to_csv('metacritic_games_cleaned.csv', index=False)
files.download('metacritic_games_cleaned.csv')
```
