```python
# Imports

import pandas as pd
import io

from tabulate import tabulate
from google.colab import files

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['metacritic_games_cleaned.csv']))
df.head()
```

```python
stats_subset = df[['metascore', 'user_score' ]].describe()
print(tabulate(stats_subset, headers=stats_subset.columns, tablefmt='fancy_grid'))
```
